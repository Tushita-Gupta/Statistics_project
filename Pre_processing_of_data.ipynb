```python
import pandas as pd
```

Pre-processing the json files and joining them to get a single dataframe and  create the response variable

```python
df_transcript = pd.read_json('C:/Users/tushi/Downloads/archive (3)/transcript.json', lines = True)
```

```python
df_transcript['event'].unique()
```

```python
df_portfolio = pd.read_json('C:/Users/tushi/Downloads/archive (3)/portfolio.json', lines = True)
```

```python
df_profile = pd.read_json('C:/Users/tushi/Downloads/archive (3)/profile.json', lines = True)
```

```python
df_transcript.drop(df_transcript.index[df_transcript['event'] == 'transaction'], inplace=True)
```

```python
df_transcript
```

```python
df_transcript['value'] = df_transcript['value'].astype(str).str[14:46]
```

```python
df_transcript
```

```python
df_portfolio
```

```python
new_df=pd.merge(df_portfolio,df_transcript,how='inner',left_on=['id'],right_on=['value'])
```

```python
new_df
```

```python
df_profile
```

```python
new_df_2=pd.merge(new_df,df_profile,how='inner',left_on=['person'],right_on=['id'])
```

```python
new_df_2
```

```python
columns = ['channels', 'id_x','id_y']
new_df_2.drop(columns, inplace=True, axis=1)
```

```python
new_df_2
```

```python
new_df_2.rename({'person': 'customer_id', 'value': 'offer_id'}, axis=1, inplace=True)
```

```python
new_df_2
```

```python
new_df_2.drop(new_df_2.index[new_df_2['offer_type'] == 'informational'], inplace=True)
```

```python
new_df_2.nunique()
```

```python
new_df_2['Customer_response']=[1 if new_df_2['event'].iloc[i]=="offer completed" else 0 for i in range(0,len(new_df_2))]
```

```python
new_df_2
```

```python
new_df_2.to_csv("C:/Users/tushi/Downloads/raw_data.csv", encoding='utf-8', index=False)
```
